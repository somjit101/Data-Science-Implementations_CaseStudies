```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mtp


irisData = pd.read_csv('Datasets/IRIS.csv')
g = irisData.groupby('species')
numG = len(g)
plotNum = 1
for gName, gDF in g:
#     print(gName)
    x = gDF['petal_width']
    y = gDF['sepal_width']
#     mtp.subplot(numG, 1, plotNum)
    mtp.scatter(x, y)
    mtp.xlabel('petal width')
    mtp.ylabel('sepal_width')
    mtp.legend()

#     print(gDF[gDF.columns[:-1]])
```

```python
#Pair Plots
import pandas as pd
import numpy as np
import matplotlib.pyplot as mtp
import seaborn as sea


irisData = pd.read_csv('Datasets/IRIS.csv')

sea.set_style("whitegrid")
sea.pairplot(irisData, hue="species", height=3)
mtp.show()
```

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as mtp
import seaborn as sea

irisData = pd.read_csv('Datasets/IRIS.csv')
mtp.figure()
sea.boxplot(x = "species", y = "petal_width", data = irisData)
mtp.grid()

mtp.figure()
sea.violinplot(x = "species", y = "petal_width", data = irisData)
mtp.grid()


```

```python
#Bivariate Analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as mtp
import seaborn as sea

irisData = pd.read_csv('Datasets/IRIS.csv')
sea.jointplot(x = "petal_length", y = "petal_width", data = irisData, kind = "kde", height = 10)
mtp.show()
```
